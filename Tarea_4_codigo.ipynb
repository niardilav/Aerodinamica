```python
#Problema 5
# --- 1. DATOS Y CONSTANTES ---
# Convertir a unidades consistentes: ft y rad
B = 3                                           # Número de aspas
R = 5.0                                         # Radio de la hélice (ft)
r = 3.0                                         # Posición radial (ft)
c = 5.65 / 12.0                                 # Cuerda (ft)
beta = np.deg2rad(47.7)                         # Ángulo de calaje (rad)
a0 = 5.73                                       # Pendiente de la curva Cl (rad^-1)
rho = 0.002377                                  # Densidad del aire (slugs/ft^3)

# Conversiones de velocidad y rpm
n = 1800 / 60.0                                 # RPS (rev/s)
V_inf = 368.0 * 5280.0 / 3600.0                 # Velocidad de avance (ft/s)

# Cálculos preliminares
Omega = 2.0 * np.pi * n                         # Velocidad angular (rad/s)
sigma = B * c / (2.0 * np.pi * r)               # Firmeza local
print(f"Omega: {Omega:.3f} rad/s")
print(f"V_inf: {V_inf:.3f} ft/s")
print(f"Sigma (σ): {sigma:.4f}")

# --- 2. ALGORITMO BEM ITERATIVO ---

# (a) Inicializar a y a'
a = 0.01
a_prime = 0.001
epsilon = 1e-3 # Criterio de convergencia
max_iterations = 100
iteration = 0

print("\n--- Inicio del Algoritmo BEM ---")

while iteration < max_iterations:
    # 1. Almacenar valores anteriores
    a_old = a
    a_prime_old = a_prime

    # 2. Calcular phi_0 (Ángulo de flujo efectivo)
    tan_phi_0 = V_inf * (1.0 + a) / (Omega * r * (1.0 - a_prime))
    phi_0 = np.arctan(tan_phi_0)

    # 3. Calcular alpha (Ángulo de ataque)
    alpha = beta - phi_0

    # 4. Obtener coeficientes aerodinámicos (Cl y Cd)
    # Perfil simétrico NACA 0015 -> alpha_L=0 = 0
    Cl = a0 * alpha # alpha está en radianes
    Cd = 0.03 * Cl
    
    # 5. Calcular los nuevos factores de inducción (a_new y a_prime_new)
    # Usando las ecuaciones de equilibrio (citadas del documento)

    # Para 'a':
    try:
        den_a = 2.0 * (1.0 - np.cos(2.0 * phi_0))
        num_a = sigma * (Cl * np.cos(phi_0) - Cd * np.sin(phi_0))
        term_a = num_a / den_a
        a_new = term_a / (1.0 - term_a)
    except ZeroDivisionError:
        print("Advertencia: División por cero para 'a'. Usando valor antiguo.")
        a_new = a_old

    # Para 'a_prime':
    try:
        den_a_prime = 2.0 * np.sin(2.0 * phi_0)
        num_a_prime = sigma * (Cl * np.sin(phi_0) + Cd * np.cos(phi_0))
        term_a_prime = num_a_prime / den_a_prime
        a_prime_new = term_a_prime / (1.0 + term_a_prime)
    except ZeroDivisionError:
        print("Advertencia: División por cero para 'a_prime'. Usando valor antiguo.")
        a_prime_new = a_prime_old

    # 6. Criterio de Convergencia
    a = a_new
    a_prime = a_prime_new
    
    # Manejo de la convergencia con divisiones por cero (para evitar errores en la condición)
    if a_old == 0:
        a_err = np.abs(a - a_old) 
    else:
        a_err = np.abs((a - a_old) / a)
    
    if a_prime_old == 0:
        a_prime_err = np.abs(a_prime - a_prime_old)
    else:
        a_prime_err = np.abs((a_prime - a_prime_old) / a_prime)

    iteration += 1

    # print(f"Iteración {iteration}: a={a:.4f}, a'={a_prime:.4f}, phi_0={np.rad2deg(phi_0):.2f}°")

    if a_err < epsilon and a_prime_err < epsilon:
        print(f"Convergencia lograda en la iteración {iteration}.")
        break

if iteration == max_iterations:
    print("Advertencia: No se alcanzó la convergencia en el número máximo de iteraciones.")

# --- 3. CÁLCULO DE RESULTADOS FINALES ---

# Factores de Inducción (a) y (a')
print("\n--- (a) Factores de Inducción ---")
print(f"a (Factor de inducción axial): {a:.4f}")
print(f"a' (Factor de inducción tangencial): {a_prime:.4f}")

# Ángulo de ataque y coeficientes finales
alpha_deg = np.rad2deg(alpha)
phi_0_deg = np.rad2deg(phi_0)
print(f"Ángulo de flujo efectivo (ϕ₀): {phi_0_deg:.2f}°")
print(f"Ángulo de ataque (α): {alpha_deg:.2f}°")
print(f"Cl: {Cl:.4f}, Cd: {Cd:.6f}")

# Velocidad efectiva (V_E)
V_E = np.sqrt((V_inf * (1.0 + a))**2 + (Omega * r * (1.0 - a_prime))**2)
print(f"Velocidad efectiva (V_E): {V_E:.3f} ft/s")

# Contribuciones por unidad de longitud (dT/dr y dM/dr)
half_rho_VE2 = 0.5 * rho * V_E**2

# dL/dr y dD/dr (fuerzas diferenciales de un solo elemento por unidad de longitud)
dL_dr_elem = Cl * half_rho_VE2 * c
dD_dr_elem = Cd * half_rho_VE2 * c

# dF_T1 y dF_M1 (componentes del empuje y fuerza de momento del elemento)
# dT1/dr y dFM1/dr

# (b) Contribución al empuje por unidad de longitud (dT/dr)
# dT/dr = B * (dT1/dr)
dT_dr = (dL_dr_elem * np.cos(phi_0) - dD_dr_elem * np.sin(phi_0))

print("\n--- (b) Contribución al Empuje (dT/dr) ---")
print(f"dT/dr: {dT_dr:.2f} lb/ft") # Las unidades son lbf/ft (libras fuerza por pie)
print(f"B*dT/dr: {B*dT_dr:.2f} lb/ft")
# (c) Contribución al momento-par por unidad de longitud (dM/dr)
# dM/dr = r * (dFM1/dr)
dFM_dr_elem = dL_dr_elem * np.sin(phi_0) + dD_dr_elem * np.cos(phi_0)
dM_dr = r * dFM_dr_elem

print("\n--- (c) Contribución al Momento-Par (dM/dr) ---")
print(f"dM/dr: {dM_dr:.2f} lb⋅ft/ft") # Las unidades son lbf*ft/ft (libras-fuerza * pie / pie)
print(f"B*dM/dr: {B*dM_dr:.2f} lb⋅ft/ft")
```

```python
#Punto 3
import numpy as np
import matplotlib.pyplot as plt

# --- 1. DATOS DE ENTRADA Y FÍSICOS ---

# Datos del Planeador
MASA = 200.0        # Masa del planeador (kg)
AREA_ALAR = 12.0    # Área alar (S, m^2)
H_INICIAL = 1200.0  # Altura de vuelo inicial (h, m)

# Constantes Físicas y ISA
G = 9.81            # Aceleración de la gravedad (m/s^2)
PESO = MASA * G     # Peso del planeador (N)
CARGA_ALAR = PESO / AREA_ALAR # W/S (N/m^2)
R_AIR = 287.05      # Constante de gas ideal del aire (J/(kg*K))
T_0 = 288.15        # Temperatura al nivel del mar (K)
P_0 = 101325.0      # Presión al nivel del mar (Pa)
LAMBDA = -0.0065    # Tasa de caída de temperatura (K/m)

# Coeficientes proporcionados (Vectores NumPy)
cl_polar = np.array([0.02185546, 0.02185546, 0.17154632, 0.62969058, 0.94268069, 1.08329883, 1.21938130, 1.30103093])
cd_perfil = np.array([0.00700001, 0.00767744, 0.00812904, 0.00993550, 0.01287098, 0.01603227, 0.02054840, 0.02506453])

# Corrección de Ala Finita (Ala Elíptica)
AR = 11.0
E_FACTOR = 1.0      # Factor de eficiencia de Oswald para ala elíptica (e = 1.0)
K = 1 / (np.pi * AR * E_FACTOR) 


# --- 2. CÁLCULO AERODINÁMICO DE LA POLAR DEL ALA ---

# C_D_ala = C_D_perfil + K * C_L^2 (Uso de NumPy)
cd_inducido = K * cl_polar**2
cd_ala = cd_perfil + cd_inducido

# Calcular la relación de planeo (E = L/D)
E_polar_corregido = cl_polar / cd_ala

# Determinar el óptimo de C_L y C_D
E_max_corregido = np.max(E_polar_corregido)
indice_optimo_corr = np.argmax(E_polar_corregido)

# Coeficientes Óptimos
C_L_OPTIMO_CORR = cl_polar[indice_optimo_corr]
C_D_OPTIMO_CORR = cd_ala[indice_optimo_corr]
C_L_MAX_CORR = cl_polar[-1] # Coeficiente de sustentación máximo


# --- 3. FUNCIONES DE DENSIDAD ISA (SOLO NUMPY) ---

def calcular_densidad_ISA(h):
    """Calcula la densidad del aire (rho) a una altitud h (en metros) usando el modelo ISA."""
    if h < 0: h = 0 
    
    T = T_0 + LAMBDA * h
    if T <= 0: return np.nan

    exponente = -G / (LAMBDA * R_AIR)
    factor_p = 1 + (LAMBDA * h / T_0)
    P = P_0 * np.power(factor_p, exponente)
    rho = P / (R_AIR * T)
    return rho

# --- 4. SIMULACIÓN E INTEGRACIÓN NUMÉRICA ---

num_segmentos = 120  
dh = H_INICIAL / num_segmentos

# Vector de alturas al inicio de cada segmento
alturas_inicio_segmento = np.linspace(H_INICIAL, dh, num_segmentos)

# Inicialización de vectores de resultados
tiempos_acumulados = np.zeros(num_segmentos + 1)
alturas_grafica = np.linspace(H_INICIAL, 0, num_segmentos + 1)
velocidades_optimas = np.zeros(num_segmentos + 1)
tiempo_total_s = 0.0

# Constante para V_optima (uso de np.sqrt)
CONST_VEL = np.sqrt((2 * CARGA_ALAR) / C_L_OPTIMO_CORR) 

# Inicializar V_optima en t=0 (H_INICIAL)
RHO_INICIAL = calcular_densidad_ISA(H_INICIAL)
V_opt_inicial_ms = CONST_VEL / np.sqrt(RHO_INICIAL)
velocidades_optimas[0] = V_opt_inicial_ms

# Bucle de integración
for i in range(num_segmentos):
    h_segmento = alturas_inicio_segmento[i]
    rho = calcular_densidad_ISA(h_segmento)

    # Velocidad Óptima al inicio del segmento (Velocidad de la Hipotenusa V)
    V_opt_h = CONST_VEL / np.sqrt(rho) 

    # Velocidad de Descenso (Vv = V * sin(theta)) 
    V_v_h = V_opt_h / E_max_corregido

    # Tiempo para recorrer este segmento
    dt = dh / V_v_h
    tiempo_total_s += dt

    tiempos_acumulados[i+1] = tiempo_total_s
    
    # Calcular V_optima al final del segmento (punto i+1)
    h_final_segmento = alturas_inicio_segmento[i] - dh
    rho_final = calcular_densidad_ISA(h_final_segmento)
    velocidades_optimas[i+1] = CONST_VEL / np.sqrt(rho_final)


# --- 5. RESULTADOS E IMPRESIÓN ---

t_min = tiempo_total_s / 60.0
segundos_restantes = (t_min % 1) * 60
minutos_enteros = int(t_min)

V_stall_inicial_ms = np.sqrt((2 * PESO) / (RHO_INICIAL * AREA_ALAR * C_L_MAX_CORR))
D_max = (E_max_corregido * H_INICIAL) / 1000.0
theta_grados = np.degrees(np.arctan(1 / E_max_corregido))


print("--- Resultados del Planeador (Modelo Completo con NumPy) ---")
print(f"**Máxima Relación L/D (E_max)**: {E_max_corregido:.2f}")
print("-------------------------------------------")

# B. Velocidad de Pérdida
print(f"**B. Velocidad para entrar en pérdida (V_stall):**")
print(f"   - **V_stall (a 1200m):** {V_stall_inicial_ms * 3.6:.2f} km/h")
print("-------------------------------------------")

# C. Máxima Distancia de Planeo
print(f"**C. Máxima distancia que puede planear (D_max):** {D_max:.2f} km")
print("-------------------------------------------")

# D. Velocidad de Descenso y Tiempo de Vuelo
print(f"**D. Tiempo de vuelo total (integrado):** {minutos_enteros} min {segundos_restantes:.0f} s")
print(f"   - **V_optima (Velocidad de la Hipotenusa, a 1200m):** {V_opt_inicial_ms:.3f} m/s")
print(f"   - **Vv (Velocidad de Descenso Vertical, a 1200m):** {V_opt_inicial_ms / E_max_corregido:.4f} m/s")
print(f"   - **Ángulo de planeo (constante):** {theta_grados:.2f}°")
print("-------------------------------------------")


# --- GRÁFICAS ---

# Gráfica 1: Altura vs. Tiempo (Perfil de Descenso)
plt.figure(figsize=(10, 5))
plt.plot(tiempos_acumulados, alturas_grafica, label='Perfil de Descenso', color='darkgreen')
plt.xlabel('Tiempo (s)')
plt.ylabel('Altura (m)')
plt.title('Gráfica 1: Perfil de Descenso (Altura vs. Tiempo)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show() 

# Gráfica 2: Velocidad del Planeador (Hipotenusa) vs. Tiempo
plt.figure(figsize=(10, 5))
plt.plot(tiempos_acumulados, velocidades_optimas, label='Velocidad Óptima (V)', color='blue')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad del Planeador (V en m/s)')
plt.title('Gráfica 2: Velocidad Óptima de la Hipotenusa vs. Tiempo')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
